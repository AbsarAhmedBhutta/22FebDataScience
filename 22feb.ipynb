```python
# sorted_df = df.groupby('quantity').apply(lambda x: x.sort_values(['quantity']))
# # print the sorted DataFrame
# sorted_df

sorted = df.sort_values(['quantity'])
sorted.to_csv('Sort.csv')
```

```python
# create a subset of the DataFrame with sales between 10 and 400

# to tackle this TypeError: '>' not supported between instances of 'str' and 'int' using:

# df['sales'] = pd.to_numeric(df['sales'], errors='coerce')
sales = df[(df['sales'] >= 10) & (df['sales'] <= 400)]

sales.to_csv('sales_between_10_and_400.csv')
```

```python
# Add a new column profit - shipping cost name it gross profit

df['gross_profit'] = df['profit']-df['shipping_cost']
df.to_csv('new_col_gross_profit.csv')
```

```python
#Find out which products not available per region


all_products = df['product_name'].unique()

grouped = df.groupby('region')

not_available_df = pd.DataFrame(columns=['region', 'not_available_products'])

for region_name, region_df in grouped:
    available = region_df['product_name'].unique()
    not_available = set(all_products) - set(available)
    region_data = {'region': region_name, 'not_available_products': not_available}
    not_available_df = pd.concat([not_available_df, pd.DataFrame([region_data])], ignore_index=True)

not_available_df.to_csv('notavailable_per_region.csv')


```
